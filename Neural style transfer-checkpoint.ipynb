```python
import torch
from torch import nn, optim
from jcopdl.callback import Callback

device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
device
```

## Import data

```python
from src import load_image
```

```python
content_path = 'content/najwa.jpg'
style_path = 'style/barli.jpg'

content = load_image(content_path).to(device)
output = load_image(content_path).to(device)
style = load_image(style_path).to(device)

output.requires_grad = True
```

```python
style.shape
```

## Training preparation

```python
from model import NeuralStyleTransfer
```

```python
model = NeuralStyleTransfer().to(device)
optimizer = optim.AdamW([output], lr=0.001)
```

## Training

### Extract feature

```python
model
```

```python
content_features = model(content, layers=["4", "8"])
style_features = model(style, layers=["4", "8"])
```

### Training loop

```python
from src import DrawStyleImage
from criterion import criterion
```

```python
max_epoch = 500

for epoch in range(1, max_epoch+1):
    output_features = model(output, layers = ['4','8'])
    loss = criterion(content_features, style_features, output_features, output_features,content_weight=1, style_weight=1e6)
    loss.backward()

    optimizer.step()
    optimizer.zero_grad()
    
    if epoch % 100 == 0:
        print(f'Epoch {epoch} | Loss {loss.item():.5f}')
        _ = DrawStyleImage(output)
```
